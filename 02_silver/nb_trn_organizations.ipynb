Loading Raw Organization Table from Bronze Catalog

```python
file_path = "/Volumes/bronze_raw_medical_data/azure_blob_storage/raw_medical_data/raw_organizations.csv"

organizations_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
```

Performing Column Renaming into Snake Case

```python
organizations_df = organizations_df.withColumnsRenamed({
    'Id':'organization_id',
    'name':'organization_name',
    'ADDRESS':'organization_address',
    'CITY':'organization_city',
    'state':'organization_state',
    'ZIP':'organization_zip',
    'LAT':'organization_lat',
    'LON':'organization_lon'
})
```

Saving the Transformed Dataframe into Silver Catalog

```python
organizations_df.write.format('delta').mode('overwrite').saveAsTable('silver_medical_data.azure_blob_storage.silver_organizations')
```
