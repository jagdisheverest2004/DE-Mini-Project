Loading Raw Payers Table from Bronze Catalog

```python
file_path = "/Volumes/bronze_raw_medical_data/azure_blob_storage/raw_medical_data/raw_payers.csv"

payers_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
```

Performing Columns Renaming in form of Snake Case

```python
payers_df = payers_df.withColumnsRenamed({
    'ADDRESS':'address',
    'CITY':'city',
    'Id':'id',
    'NAME':'name',
    'PHONE':'phone',
    'STATE_HEADQUARTERED':'state_headquartered',
    'ZIP':'zip',
})
```

Saving Transformed Dataframe into Silver Catalog

```python
payers_df.write.format('delta').mode('overwrite').saveAsTable('silver_medical_data.azure_blob_storage.silver_payers')
```
