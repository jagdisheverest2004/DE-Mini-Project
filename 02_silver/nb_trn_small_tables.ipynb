```python
file_path = "/Volumes/bronze_raw_medical_data/azure_blob_storage/raw_medical_data/raw_data_dictionary.csv"

data_dictionary_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

data_dictionary_df.write.format('delta').option('mergeSchema','true').mode('overwrite').saveAsTable('silver_medical_data.azure_blob_storage.data_dictionary')
```
