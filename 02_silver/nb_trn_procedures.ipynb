```python
file_path = "/Volumes/bronze_raw_medical_data/azure_blob_storage/raw_medical_data/raw_procedures.csv"

procedures_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
```

```python
procedures_df = procedures_df.withColumnsRenamed({
    "START": "start",
    "stop": "stop",
    "PATIENT": "patient_id",
    "CODE": "code",
    "DESCRIPTION": "description",
    "BASE_COST": "base_cost",
    "REASONCODE": "reason_code",
    "REASONDESCRIPTION": "reason_description"
})
```

```python
from pyspark.sql.functions import regexp_replace

procedures_df = procedures_df.withColumns({
    "code": procedures_df.code.cast('long'),
    "base_cost": regexp_replace(procedures_df.base_cost, ',', '').cast('float')
})
```

```python
procedures_df.write.format('delta').mode('overwrite').option("mergeSchema","true").saveAsTable('silver_medical_data.azure_blob_storage.procedures')
```
