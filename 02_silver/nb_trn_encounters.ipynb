Importing Libraries

```python
from pyspark.sql import functions as F
```

Loading Encounters tables from Bronze Catalog

```python
file_path = "/Volumes/bronze_raw_medical_data/azure_blob_storage/raw_medical_data/raw_encounters.csv"

encounters_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
```

Performing Columns Renaming into Snake Case

```python
encounters_df = encounters_df.withColumnsRenamed(
    {
        'Id':'id',
        'start':'start',
        'STOP':'stop',
        'PATIENT':'patient_id',
        'organization':'organization_id',
        'PAYER':'payer_id',
        'ENCOUNTER_CLASS':'encounter_class',
        'CODE':'code',
        'DESCRIPTION':'description',
        'BASE_ENCOUNTER_COST':'base_encounter_cost',
        'TOTAL_CLAIM_COST':'total_claim_cost',
        'PAYER_COVERAGE':'payer_coverage',
        'REASONCODE':'reason_code',
        'REASONDESCRIPTION':'reason_description'
    }
)
```

Performing Data Type Casting for Encounters Dataframe

```python
encounters_df = encounters_df.withColumns({
    'base_encounter_cost':encounters_df.base_encounter_cost.cast('double'),
    'total_claim_cost':encounters_df.total_claim_cost.cast('double'),
    'payer_coverage':encounters_df.payer_coverage.cast('double'),
    'reason_code':encounters_df.reason_code.cast('double'),
    'code':encounters_df.code.cast('long')
})
```

Saving Encounters Dataframe into Silver Catalog

```python
encounters_df.write.format('delta').mode('overwrite').saveAsTable('silver_medical_data.azure_blob_storage.silver_encounters')
```
