```python
encounters_df = spark.read.table('bronze_medical_data.azure_blob_storage.encounters')
display(encounters_df)
```

```python
encounters_df = encounters_df.select(
    'Id', 'start',
    'STOP', 'PATIENT', 'organization', 'PAYER', 'ENCOUNTER_CLASS','CODE','DESCRIPTION',
    'BASE_ENCOUNTER_COST', 'TOTAL_CLAIM_COST',
    'PAYER_COVERAGE','REASONCODE','REASONDESCRIPTION'
)

display(encounters_df)
```

```python
encounters_df = encounters_df.withColumnsRenamed(
    {
        'Id':'encounter_id',
        'start':'encounter_start',
        'STOP':'encounter_stop',
        'PATIENT':'patient_id',
        'organization':'organization_id',
        'PAYER':'payer_id',
        'ENCOUNTER_CLASS':'encounter_class',
        'CODE':'code',
        'DESCRIPTION':'description',
        'BASE_ENCOUNTER_COST':'base_encounter_cost',
        'TOTAL_CLAIM_COST':'total_claim_cost',
        'PAYER_COVERAGE':'payer_coverage',
        'REASONCODE':'reason_code',
        'REASONDESCRIPTION':'reason_description'
    }
)
display(encounters_df)
```

```python
encounters_df.write.format('delta').mode('overwrite').saveAsTable('silver_medical_data.azure_blob_storage.encounters')
```
