```python
%sql
CREATE OR REPLACE TABLE gold_medical_data.azure_blob_storage.fact_main AS
with encounters as (
  select id as encounter_id,base_encounter_cost,total_claim_cost,payer_coverage from silver_medical_data.azure_blob_storage.encounters
),
procedures as (
  select encounter_id,base_cost as base_procedure_cost from silver_medical_data.azure_blob_storage.procedures
)
select encounters.*,procedures.base_procedure_cost from encounters join procedures on encounters.encounter_id=procedures.encounter_id;
```
