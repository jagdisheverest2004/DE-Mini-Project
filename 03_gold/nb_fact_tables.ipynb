```python
%sql
create or replace table gold_medical_data.fact_tables.fact_encounters as
select
  id as encounter_id,
  patient_id,
  payer_id,
  organization_id,
  base_encounter_cost,
  total_claim_cost,
  payer_coverage,
  start as encounter_start,
  stop as encounter_stop
from
  silver_medical_data.azure_blob_storage.encounters;
```

```python
%sql
create or replace table gold_medical_data.fact_tables.fact_procedures as
select
  encounter_id,
  patient_id,
  base_cost as base_procedure_cost,
  start as procedure_start,
  stop as procedure_stop
from
  silver_medical_data.azure_blob_storage.procedures;
```
