```python
%sql
create or replace table gold_medical_data.azure_blob_storage.fact_encounters as 
select id as encounter_id,base_encounter_cost,total_claim_cost,payer_coverage from silver_medical_data.azure_blob_storage.encounters;
```

```python
%sql
create or replace table gold_medical_data.azure_blob_storage.fact_procedures as 
select encounter_id,base_cost as base_procedure_cost from silver_medical_data.azure_blob_storage.procedures;
```

```python
%sql
select
  encounter_id,
  count(*)
from
  gold_medical_data.azure_blob_storage.fact_encounters
group by
  encounter_id
order by
  encounter_id;
```

```python
%sql
select
  patient_id,
  encounter_id,
  code,
  reason_code,
  count(*)
from
  silver_medical_data.azure_blob_storage.procedures
group by
  patient_id,
  encounter_id,
  code,
  reason_code
having count(*)>1
order by
  patient_id,
  encounter_id,
  code,
  reason_code
;
```
